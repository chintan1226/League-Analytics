```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data_table = pd.read_csv("lec_playerdata.csv")

dataset = np.array(data_table)

data_table
```

```python
print(dataset)
```

```python
from sklearn import linear_model

kda = []
win_p = []

for data in dataset:
    kda.append([data[8]])
    win_p.append(data[4])

x_train = kda[0:78]
y_train = win_p[0:78]
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
linear.score(x_train, y_train)
predicted= linear.predict(x_train)

plt.scatter(x = x_train, y = y_train)
plt.plot(x_train,predicted)
print('X: KDA, Y:Win_P')
```
